# SOTA trends

```python
df_release_dates = pd.read_csv("https://raw.githubusercontent.com/forecastingresearch/forecastbench/refs/heads/main/src/leaderboard/model_release_dates.csv")
df_release_dates.head()
```

```python
# Update model names to be consistent with release dates
df_tournament_leaderboard = pd.read_csv("https://raw.githubusercontent.com/forecastingresearch/forecastbench-datasets/refs/heads/main/leaderboards/csv/leaderboard_tournament.csv")
df_baseline_leaderboard = pd.read_csv("https://raw.githubusercontent.com/forecastingresearch/forecastbench-datasets/refs/heads/main/leaderboards/csv/leaderboard_baseline.csv")
for df_temp in [df_tournament_leaderboard, df_baseline_leaderboard]:
    df_temp["Model"] = df_temp["Model"].str.replace("crowd forecast", "freeze values")

# Update column name in df_release_dates to be consistent
# with leaderboards
df_release_dates = df_release_dates.rename(columns={"model": "Model"})
```

```python
def get_sota_df(df_leaderboard, df_release_dates, score_col="Overall"):
    if score_col not in df_leaderboard.columns:
        raise ValueError(f"Score column '{score_col}' not found in leaderboard dataframe.")
    
    # Merge in release dates
    df_merged = pd.merge(df_leaderboard, df_release_dates,
                        on="Model",
                        how="left")
    df_merged = df_merged.sort_values(by=score_col, ascending=True)

    # Calculate best model scores for each release date
    df_best_by_date = df_merged.groupby("model_release_date").first().reset_index()
    df_best_by_date = df_best_by_date.sort_values(by="model_release_date")

    # Calculate cumulative minimum to identify SOTA models
    df_best_by_date["cummin_score"] = df_best_by_date[score_col].cummin()
    df_best_by_date["previous_cummin"] = df_best_by_date["cummin_score"].shift(1, fill_value=np.inf)

    # Filter to SOTA models (strictly improving)
    mask = df_best_by_date[score_col] < df_best_by_date["previous_cummin"]
    df_sota = df_best_by_date[mask].copy()

    # For SOTA models, sota_score equals their actual score
    df_sota["sota_score"] = df_sota[score_col]

    # Calculate for trendline calculation
    df_sota['years_since_first_model'] = ((pd.to_datetime(df_sota['model_release_date'])
                                            - pd.to_datetime(df_sota['model_release_date'].min()))\
                                            .dt\
                                            .days / 365.25)
    
    return df_sota[["Model", 
                    "model_release_date", 
                    score_col, 
                    "sota_score", 
                    "years_since_first_model"]]


def get_sota_trendline(df_leaderboard, df_release_dates, score_col="Overall"):
    df_sota = get_sota_df(df_leaderboard, df_release_dates, score_col=score_col)
    results = smf.ols('sota_score ~ years_since_first_model', data=df_sota).fit()
    mask = df_leaderboard["Model"] == "Superforecaster median forecast"
    supers_score = df_leaderboard.loc[mask, score_col].values[0]
    predicted_parity_days_since_first_model = ((supers_score - results.params['Intercept']) 
                                               / results.params['years_since_first_model'] * 365.25)
    predicted_parity_date = (pd.to_datetime(df_sota['model_release_date'].min()) 
                         + pd.DateOffset(days=predicted_parity_days_since_first_model))
    return {"results": results, 
            "df_sota": df_sota,
            "supers_score": supers_score,
            "predicted_parity_date": predicted_parity_date}
```

```python
sota_tournament_overall = get_sota_trendline(df_leaderboard = df_tournament_leaderboard, 
                                             df_release_dates = df_release_dates, 
                                             score_col="Overall")
print(sota_tournament_overall["results"].summary())
```

```python
def get_rank_reversals(df, df_leaderboard, question_type, models, min_overlap):
    """
    Find ranking reversals between head-to-head (H2H) comparisons
    and leaderboard rankings.

    Args:
        df: DataFrame with forecasts
        df_leaderboard: Dataframe with the leaderboard
        question_type: 'market' or 'dataset'
        models: List of models to compare. These models are compared
            to every model in df_leaderboard["model"].unique()
        min_overlap: Minimum overlapping questions

    Returns:
        DataFrame with ranking reversals
    """
    # Filter data by question type
    mask = (df["market_question"] == (question_type == "market"))
    df_filtered = df[mask].copy()
    all_models = df_leaderboard["Model"].unique()
    all_models_df = df["model"].unique()

    # Get leaderboard scores
    if question_type == "market":
        score_col = "Market"
    elif question_type == "dataset":
        score_col = "Dataset"
    else:
        raise ValueError(f"Unexpected question_type={question_type} provided. "
                         f"Only market or dataset are allowed.")
    model_scores = df_leaderboard.set_index('Model')[score_col].to_dict()

    # Loop across models and collect reversals
    reversals = []
    for model_a in models:
        for model_b in all_models:
            if model_a == model_b:
                continue
            if model_a not in all_models_df:
                raise ValueError(f"{model_a} not present in df.")
            if model_b not in all_models_df:
                raise ValueError(f"{model_b} not present in df.") 

            # Get raw Brier scores
            df_a = df_filtered[df_filtered["model"] == model_a][["question_id", "brier_score"]]
            df_b = df_filtered[df_filtered["model"] == model_b][["question_id", "brier_score"]]
            df_overlap = pd.merge(df_a, df_b, 
                                how="inner", 
                                on="question_id", 
                                suffixes=("_a", "_b"))
            
            if len(df_overlap) < min_overlap:
                continue

            # Head-to-head & leaderboard comparison
            head_to_head_diff = (
                df_overlap['brier_score_a'].mean() - df_overlap['brier_score_b'].mean()
            )
            leaderboard_diff = model_scores[model_a] - model_scores[model_b]

            # Ranking reversal: signs don't match.
            # For both rankings, lower scores are better
            if (head_to_head_diff * leaderboard_diff) < 0:
                h2h_winner = model_a if head_to_head_diff < 0 else model_b
                lb_winner = model_a if leaderboard_diff < 0 else model_b

                reversals.append({
                    "model_1": model_a,
                    "model_2": model_b,
                    "question_type": question_type,
                    "n_overlap": len(df_overlap),
                    "head_to_head_diff": head_to_head_diff,
                    "leaderboard_diff": leaderboard_diff,
                    "discrepancy": abs(head_to_head_diff - leaderboard_diff),
                    "head_to_head_winner": h2h_winner,
                    "leaderboard_winner": lb_winner
                })

    return pd.DataFrame(reversals)
```

```python
df_dataset_reversals = get_rank_reversals(df=df, 
                                          df_leaderboard=df_tournament_leaderboard, 
                                          question_type="dataset", 
                                          models=top_models, 
                                          min_overlap=1000)
print(len(df_dataset_reversals))
df_dataset_reversals.head()
```

```python
df_market_reversals = get_rank_reversals(df=df, 
                                          df_leaderboard=df_tournament_leaderboard, 
                                          question_type="market", 
                                          models=top_models, 
                                          min_overlap=100)
print(len(df_market_reversals))
df_market_reversals.head()
```
