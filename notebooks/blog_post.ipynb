# GPT-4.5 market questions

```python
mask = (df["model"] == "GPT-4.5-Preview-2025-02-27 (zero shot with freeze values)") & (df['market_question'])
df_temp = df[mask].copy()
df_temp.head()
```

```python
df_temp["market_value_on_freeze_date"] = df_temp["market_value_on_freeze_date"].astype(float)
df_temp["forecast_market_delta"] = (df_temp["market_value_on_freeze_date"]  - df_temp["forecast"]).abs()
df_temp["forecast_market_delta"].describe()
```

```python
(df_temp["forecast_market_delta"].median() * 100).round(2)
```
